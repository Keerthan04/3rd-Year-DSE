```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('customer_attrition.csv')
df.head()
```

# Preprocessing

```python
#Checking for any null values
df.isnull().sum()
```

```python
### Drop irrelevant columns
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)#as not required for the apporximation
df.head()
```

```python
#viewing the metadata
df.describe()
```

```python
#Checking for the categorical and non categorical columns
category = []
non_category =[]
for col in df.columns:
    if(df[col].dtype=='O'):
        category.append(col)
    else:
        non_category.append(col)
print('Category columns are ',category)
print('Non Categorical Columns are ',non_category)
```

```python
#viewing the Gender column
df['Gender'].value_counts()
#has 2 categories
```

```python
#Viewing the Geography Column
df['Geography'].value_counts()
#has 3 categories
```

```python
## Encode categorical variables
label_encoder_gender=LabelEncoder()
df['Gender']=label_encoder_gender.fit_transform(df['Gender'])
df.head()
```

```python
df['Gender'].value_counts()
```

```python
## Onehot encode 'Geography
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo=OneHotEncoder()
geo_encoder=onehot_encoder_geo.fit_transform(df[['Geography']]).toarray()
geo_encoder
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df=pd.DataFrame(geo_encoder,columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df.head()
```

```python
## Combine one hot encoder columns with the original data
df=pd.concat([df.drop('Geography',axis=1),geo_encoded_df],axis=1)
df.head()
```

```python
## DiVide the dataset into indepent and dependent features
X=df.drop('Exited',axis=1)
y=df['Exited']

## Split the data in training and tetsing sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=42)

## Scale these features
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
y_train
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
(X_train.shape[1],)
```

```python
models_dict = [
    Sequential([
        Input(shape=(X_train.shape[1],),name='input'),
        Dense(128, activation='relu', name='model1_h1'),
        Dense(1, activation='sigmoid', name='model1_output'),
    ]),
    Sequential([
        Input(shape=(X_train.shape[1],),name='input'),
        Dense(128, activation='relu', name='model2_h1'),
        Dense(64, activation='relu', name='model2_h2'),
        Dense(1, activation='sigmoid', name='model2_output'),
    ]),
    Sequential([
        Input(shape=(X_train.shape[1],),name='input'),
        Dense(128, activation='relu', name='model3_h1'),
        Dense(64, activation='relu', name='model3_h2'),
        Dense(32, activation='relu', name='model3_h3'),
        Dense(1, activation='sigmoid', name='model3_output'),
    ])]
models_dict
```

```python
for model in models_dict:
    model.summary()
```

```python
batch_size = 32
epochs = [10, 20, 30]
learning_rate = [0.1, 0.01, 0.001]
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import Recall, Precision

%matplotlib inline

# Define hyperparameters
batch_size = 32
epochs = [10, 20, 30]
learning_rates = [0.1, 0.01, 0.001]

performance = []

def model_train(models):
    for index, model in enumerate(models):
        for epoch in epochs:
            for lr in learning_rates:
                print(f'Compiling Model{index+1} with epoch {epoch} and Learning Rate {lr}')
                print('-----------------------------------------------------')
                model.compile(
                    optimizer=Adam(learning_rate=lr),
                    loss='binary_crossentropy',
                    metrics=['accuracy', Precision(name='precision'), Recall(name='recall')]  # Ensure these metrics are used
                )
                print('Compile of Model Complete')
                
                print(f'Fitting the Model{index+1} with epoch {epoch} and Learning Rate {lr}')
                print('------------------------------------------------------')
                
                fitted_model = model.fit(
                    X_train, y_train,
                    epochs=epoch,
                    batch_size=batch_size,
                    validation_split=0.15
                )
                print('Fit of Model Complete')
                
                print(f'Evaluating the Model{index+1} with epoch {epoch} and Learning Rate {lr}')
                loss, accuracy, precision, recall = model.evaluate(X_test, y_test, batch_size=batch_size)
                
                train_accuracy = fitted_model.history.get('accuracy', [])
                train_precision = fitted_model.history.get('precision', [])
                train_recall = fitted_model.history.get('recall', [])
                epochs_range = range(1, epoch+1)
                
                print(f'Plotting the Model{index+1} Train and Test Accuracy with respect to Epochs')
                plt.figure(figsize=(12, 6))
                plt.subplot(1, 2, 1)
                if train_accuracy and fitted_model.history.get('val_accuracy', []):
                    plt.plot(epochs_range, train_accuracy, 'g', label='Training Accuracy')
                    plt.plot(epochs_range, fitted_model.history.get('val_accuracy', []), 'b', label='Validation Accuracy')
                plt.xlabel('Epochs')
                plt.ylabel('Accuracy')
                plt.title(f'Model{index+1} Accuracy')
                plt.legend()
                
                print(f'Plotting the Model{index+1} Train and Test Loss with respect to Epochs')
                plt.subplot(1, 2, 2)
                if fitted_model.history.get('loss', []) and fitted_model.history.get('val_loss', []):
                    plt.plot(epochs_range, fitted_model.history['loss'], 'g', label='Training Loss')
                    plt.plot(epochs_range, fitted_model.history.get('val_loss', []), 'b', label='Validation Loss')
                plt.xlabel('Epochs')
                plt.ylabel('Loss')
                plt.title(f'Model{index+1} Loss')
                plt.legend()
                
                plt.show()
                
                # Append the result
                performance.append({
                    'model_name': f'Model{index+1}',
                    'Epochs': epoch,
                    'Learning Rate': lr,
                    'Train Accuracy': train_accuracy[-1] if train_accuracy else None,
                    'Test Accuracy': accuracy,
                    'Train Precision': train_precision[-1] if train_precision else None,
                    'Test Precision': precision,
                    'Train Recall': train_recall[-1] if train_recall else None,
                    'Test Recall': recall,
                    'Train Loss': fitted_model.history['loss'][-1] if fitted_model.history['loss'] else None,
                    'Test Loss': loss
                })
```

```python
model_train(models_dict)
```

```python
performance_df = pd.DataFrame(performance)
performance_df
```
